# PMPL Testing Notebook

This notebook tests the functionality of the pmpl (Python Matplotlib Plotting Library) package.

## Features tested:
- Context managers for temporary style application
- Direct axis formatting functions
- Preset styles (base, horizontal, vertical)
- Style switching and comparison

## 1. Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np

import pmpl

print("pmpl imported successfully!")
```

## 2. Test Base Style Formatting

```python
# Create sample data
categories = ["Category A", "Category B", "Category C", "Category D"]
values = [23, 45, 56, 78]

# Create plot with base formatting
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(categories, values, color="steelblue")
ax.set_title("Base Style Formatting Test")
ax.set_ylabel("Values")

# Apply base formatting
pmpl.format_base(ax)

plt.tight_layout()
plt.show()
```

## 3. Test Horizontal Style with Context Manager

```python
with pmpl.style("horizontal"):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.barh(categories, values, color="coral")
    ax.set_title("Horizontal Style with Context Manager")
    ax.set_xlabel("Values")
    plt.tight_layout()
    plt.show()

# Verify style reverts after context
print("Context exited - styles should revert to defaults")
```

## 4. Test Vertical Style with Context Manager

```python
# Test with line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

with pmpl.style("vertical"):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(x, y, linewidth=2, color="seagreen")
    ax.set_title("Vertical Style - Line Plot")
    ax.set_xlabel("X values")
    ax.set_ylabel("Y values")
    plt.tight_layout()
    plt.show()
```

```python
# Test with vertical bar chart
with pmpl.style("vertical"):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(categories, values, color="mediumpurple")
    ax.set_title("Vertical Style - Bar Chart")
    ax.set_ylabel("Values")
    plt.tight_layout()
    plt.show()
```

## 5. Test Direct Axis Formatting

```python
# Create plot without context manager
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(categories, values, color="tomato")
ax.set_title("Direct Horizontal Formatting")
ax.set_xlabel("Values")

# Apply formatting directly to axis
pmpl.format_horizontal(ax)

plt.tight_layout()
plt.show()
```

```python
# Test vertical formatting directly
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y * 2, linewidth=2, color="dodgerblue")
ax.set_title("Direct Vertical Formatting")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")

# Apply formatting directly
pmpl.format_vertical(ax)

plt.tight_layout()
plt.show()
```

## 6. Test Style Switching

```python
# Create multiple subplots with different styles
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Subplot 1: Horizontal style
axes[0, 0].barh(categories, values, color="coral")
axes[0, 0].set_title("Horizontal Formatting")
pmpl.format_horizontal(axes[0, 0])

# Subplot 2: Vertical style
axes[0, 1].bar(categories, values, color="steelblue")
axes[0, 1].set_title("Vertical Formatting")
pmpl.format_vertical(axes[0, 1])

# Subplot 3: Base style
axes[1, 0].scatter(range(len(categories)), values, s=200, color="seagreen")
axes[1, 0].set_xticks(range(len(categories)))
axes[1, 0].set_xticklabels(categories, rotation=45)
axes[1, 0].set_title("Base Formatting")
pmpl.format_base(axes[1, 0])

# Subplot 4: Line plot with vertical
axes[1, 1].plot(x[:50], y[:50], linewidth=2, color="purple")
axes[1, 1].set_title("Line Plot - Vertical Style")
pmpl.format_vertical(axes[1, 1])

plt.tight_layout()
plt.show()
```

## 7. Compare Styles Side-by-Side

```python
# Same data, different styles for comparison
comparison_data = {"Q1": 15, "Q2": 28, "Q3": 42, "Q4": 35}

quarters = list(comparison_data.keys())
revenues = list(comparison_data.values())

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# All three plots with same data
for i, ax in enumerate(axes):
    ax.bar(quarters, revenues, color=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"])
    ax.set_title(
        ["Base Style", "Horizontal Grid Style", "Vertical Grid Style"][i], fontsize=14
    )
    ax.set_ylabel("Revenue ($M)")
    ax.set_xlabel("Quarter")

# Apply different styles
pmpl.format_base(axes[0], grid_axis="both")
pmpl.format_base(axes[1], grid_axis="x")
pmpl.format_vertical(axes[2])

plt.suptitle("Style Comparison: Same Data, Different Formatting", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
